In this section we will explore the bag of words model and lean how to encode text using this model.

A bag-of-words is a representaion of text that describes the occurrence of wirds within a document.


```python
# Bag of words with sklearn
from sklearn.feature_extraction.text import TfidfVectorizer

#list of the text decuments
text = ["The quick brown fox jumped over the lazy dog.",
        "The dog.",
        "The fox."]

# create the transform
vectorizer = TfidfVectorizer()

#tokenize and build vocab
vectorizer.fit(text)

# summarize
print(vectorizer.vocabulary_)
print(vectorizer.idf_)

# encode document
vector = vectorizer.transform([text[0]])

# summarize encoded vector
print(vector.shape)
print(vector.toarray())
```

```python
# bag of words keras
from keras.preprocessing.text import  Tokenizer
# define 5 documents
docs = ['Well done!',
        'Good work',
        'Great effort',
        'nice work',
        'Excellent']

# creat the tokenizer
t = Tokenizer()

# fit the tokenizer on the documents 
t.fit_on_texts(docs)

#summarize what was learned
print(t.word_counts)
print(t.document_count)
print(t.word_index)
print(t.word_docs)
```
